```python
import seaborn as sns
import matplotlib.pyplot as plt

# Given data
papers = [
    [2019, 'Research paper'],
    [2021, 'Research paper'],
    [2021, 'Research paper'],
    [2015, 'Research paper'],
    [2022, 'Research paper'],
    [2020, 'Research paper'],
    [2020, 'Case study'],
    [2018, 'Research paper'],
    [2020, 'Survey paper'],
    [2022, 'Research paper']
]

# Extracting x and hue data from the given data
years = [paper[0] for paper in papers]
paper_types = [paper[1] for paper in papers]

# Creating a histogram using seaborn
sns.histplot(data=papers, x=years, hue=paper_types, palette='viridis', multiple='stack')

# Setting x-axis and y-axis labels
plt.xlabel('Year')
plt.ylabel('Number of Papers')

# Displaying the plot
plt.show()

```
